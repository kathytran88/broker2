```python
from selenium import webdriver
from bs4 import BeautifulSoup 
import time 

driver = webdriver.Chrome()
driver.get('https://www.bizquest.com/alabama-business-brokers/page-3/')
time.sleep(10)

page_source = driver.page_source 
soup = BeautifulSoup(page_source, 'html.parser')

## Dictionary ##
brokers = []

locations = soup.find_all('div', {'class': 'row broker'})

for location in locations:
    broker = {
        'company': 'N/A',
        'name': 'N/A',
        'state': 'N/A',
        'link': 'N/A'
    }

    # Link
    try:
        link_loc1 = location.find('div', {'class': 'col-xs-12 col-sm-6 brokerCard'})
        link_loc2 = link_loc1.find('a', {'class': "bnw-result"})
        link = link_loc2['href']
        broker['link'] = link
    except:
         print('No url found')

    # Name 
    try:
         name_loc1 = location.find('div', {'class': 'col-xs-12 col-sm-6 brokerCard'})
         name_loc2 = name_loc1.find('div', {'class': 'brokerInfo'})
         name_loc3 = name_loc2.find('a', {'class': 'bnw-result'})
         name_loc4 = name_loc3.find('p', {'class': 'brokerName'})
         name = name_loc4.contents[0].strip()
         broker['name'] = name
    except:
         print('No name found')

    # State & Company
    try:
         state_loc1 = location.find('div', {'class': 'col-xs-12 col-sm-6 brokerCompany'})
         state_loc2 = state_loc1.find('p', {'class': 'brokerage'})
         company = state_loc2.contents[0].strip()
         state_loc3 = state_loc2.find('span', {'class': 'brokerLocation'})
         state = state_loc3.get_text(strip=True)

         broker['company'] = company 
         broker['state'] = state
    except:
         print('No state and company found')

    # Add 
    brokers.append(broker)

driver.quit()
```

```python
print(brokers)
```

```python
import pandas as pd
df = pd.DataFrame(brokers)
df.head()
```

# Phone number by accessing all urls

```python
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Initialize the WebDriver
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Run in headless mode
service = Service('path/to/chromedriver')  # Path to chromedriver

driver = webdriver.Chrome(service=service, options=options)

phones = []

for link in df['link']:
    driver = webdriver.Chrome()
    driver.get(link)
    time.sleep(10)

    page = driver.page_source 
    soup = BeautifulSoup(page, 'html.parser')

    # Find element 
    phone_loc = soup.find('span', {'id': 'phone'})
    phone = phone_loc.get_text(strip=True)

    phones.append(phone)

driver.quit()

print(phones)
```

```python
df['phone'] = phones
df.head()
```
