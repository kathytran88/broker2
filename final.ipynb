```python
import pandas as pd 
import numpy as np

df = pd.read_excel('company.xlsx')
del df['Unnamed: 0']
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.rename(columns = {'state': 'location'})
df.head()
```

```python
state = []
city = []

for place in df['location']:
    location = place.split(',')
    if len(location) == 1:
        city.append(np.nan)
        state.append(location[0])
    else:
        city.append(location[0])
        state.append(location[1])
print(len(state))
print(len(city))
```

```python
df.shape
```

```python
df['state'] = state 
df['city'] = city 
df.head()
```

```python
df.duplicated().sum()
```

```python
df = df.sort_values(by='name')
df.head()
```

```python
name_dup = df['name'].duplicated(keep=False)
name_dup_df = df[name_dup]
display(name_dup_df)
```

```python
df = df.drop_duplicates(subset=['phone number'], keep='first')
df.shape
```

```python
first = []
last = []

for name in df['name']:
    name_split = name.split(' ')
    if len(name_split) == 2:
        first.append(name_split[0])
        last.append(name_split[1])

    else:
        first.append(np.nan)
        last.append(np.nan)

print(len(first))
print(len(last))
```

```python
df['first name'] = first 
df['last name'] = last
df.head()
```

```python
df = df[['company', 'first name', 'last name', 'name', 'phone number', 'location', 'city', 'state', 'link']]
df.head()
```

```python
df = df.reset_index(drop=True)
df.head()
```

```python
df = df.rename(columns = {'company': 'Company' , 'first name': 'First Name', 'last name': 'Last Name', 'name': 'Full Name', 'phone': 'Corporate Phone', 'location': 'Location', 'city': 'City', 'state': 'State', 'link': 'Profile Link'})
df.head()
```

```python
df.shape
```

```python
df.to_excel('official.xlsx')
```
