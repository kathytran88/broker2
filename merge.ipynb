```python
import pandas as pd 
import numpy as np
import glob 

excel_files = glob.glob('./files/*.xlsx')

all_df = []

for file in excel_files:
    dfs = pd.read_excel(file)
    all_df.append(dfs)

df = pd.concat(all_df, ignore_index = True)
df.head()
```

```python
del df['Unnamed: 0']
df.shape
```

```python
df.isnull().sum()
```

```python
phone_null = df[df['phone'].isnull()]
display(phone_null)
```

```python
for link in phone_null['link']:
    print(link)
```
