```python
import pandas as pd 
import numpy as np

df = pd.read_excel('official3.xlsx')
del df['Unnamed: 0']
df.head()
```

```python
for num in range(len(df)):
    phone = df.loc[num, 'phone number']
    if pd.notnull(phone):  # Check if the value is not null
        phone_str = str(phone)
        if len(phone_str) != 10:
            print(f"Name: {df.loc[num, 'Full Name']}, Phone: {phone}")
```

```python
df = df.rename(columns = {'phone number': 'Corporate Phone'})
```

```python
for num in range(len(df)):
    phone = df.loc[num, 'Corporate Phone']
    phone = str(phone)
```

```python
for c in range(len(df['Corporate Phone'])):
    phone = df.loc[c, 'Corporate Phone']
    if pd.notnull(phone) and len(str(phone)) == 10:
        phone_str = str(phone)
        formatted_phone = '+1 ' + phone_str[:3] + '-' + phone_str[3:6] + '-' + phone_str[6:]
        df.loc[c, 'Corporate Phone'] = formatted_phone
```

```python
df = df.sort_values(by='State')
df = df.reset_index(drop=True)
```
