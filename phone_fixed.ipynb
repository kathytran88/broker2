```python
import pandas as pd 
import numpy as np

df = pd.read_excel('phonenull.xlsx')
del df['Unnamed: 0']
df.head()
```

```python
df.isnull().sum()
```

```python
phone_str = ''

for phone in df['phone']:
    for character in phone:
        if character.isnumeric() == True:
            phone_str += character 
    phone_str += ' '

print(phone_str)
```

```python
phone_list = phone_str.split(' ')
print(phone_list)
```

```python
phone_list2 = []

for each in phone_list:
    if len(each) == 11:
        new = each[1:]
        phone_list2.append(new)
    else:
        phone_list2.append(each)

print(phone_list2)
```

```python
for each in phone_list2:
    if len(each) != 10:
        print(each)
```

```python
del phone_list2[len(phone_list2) - 1]
```
