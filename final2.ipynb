```python
import pandas as pd 
import numpy as np

df = pd.read_excel('official.xlsx')
del df['Unnamed: 0']
df.head()
```

```python
for num in range(len(df)):
    the_name = df['Full Name'][num]
    if ',' in the_name:
        name_split = the_name.split(',')
        df.loc[num, 'Full Name'] = name_split[0]
```

```python
for num in range(len(df)):
    first_name = df.loc[num, 'First Name']
    last_name = df.loc[num, 'Last Name']
    full_name = df.loc[num, 'Full Name']

    if pd.isnull(first_name) and pd.isnull(last_name):
        full_name_split = full_name.split(' ')
        if len(full_name_split) == 2:
            df.loc[num, 'First Name'] = full_name_split[0]
            df.loc[num, 'Last Name'] = full_name_split[1]
        elif len(full_name_split) == 3:
            df.loc[num, 'First Name'] = full_name_split[0] + ' ' + full_name_split[1]
            df.loc[num, 'Last Name'] = full_name_split[-1]
```

```python
df.to_excel('official2.xlsx')
```
